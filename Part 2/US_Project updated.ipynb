```python
def format_quarters(x, pos):
    date = mdates.num2date(x)  # Convert matplotlib date to datetime
    quarter = (date.month - 1) // 3 + 1  # Calculate the quarter
    if quarter == 1:  # Show year only for Q1
        return f"{date.year}-Q{quarter}"
    else:  # Show only the quarter for Q2, Q3, Q4
        return f"Q{quarter}"
```

```python
# Define date range
start_date = datetime.datetime(1988, 1, 1)
end_date = datetime.datetime(2024, 12, 31)
```

```python
# Download CPI data
df_cpi = web.DataReader('CPIAUCSL', 'fred', start_date, end_date)
df_cpi.dropna(inplace=True)

# Create a column for quarterly CPI using 3-month average (only at quarter-end months)
df_cpi['Quarterly CPI'] = df_cpi['CPIAUCSL'].rolling(window=3).mean()

# Keep only the months that are quarter-ends: March, June, September, December
df_cpi_quarterly = df_cpi[df_cpi.index.month.isin([3, 6, 9, 12])].copy()

# Compute annualized inflation rate
df_cpi_quarterly['Annual Inflation Rate'] = 400 * np.log(
    df_cpi_quarterly['Quarterly CPI'] / df_cpi_quarterly['Quarterly CPI'].shift(1)
)

df_unemp = web.DataReader('UNRATE', 'fred', start_date, end_date)
df_unemp_quarterly = df_unemp.resample('QE').mean()
df_unemp_quarterly.rename(columns={'UNRATE': 'Unemployment Rate Quarterly'}, inplace=True)
df_unemp_quarterly

df_interest = web.DataReader('GS10', 'fred', start_date, end_date)
df_interest_quarterly = df_interest.resample('QE').mean()
df_interest_quarterly.rename(columns={'GS10': 'Interest Rate Quarterly'}, inplace=True)
df_interest_quarterly

df_gdp = web.DataReader('GDPC1', 'fred', start_date, end_date)
# Compute quarterly GDP change (log difference * 100)
df_gdp['Quarterly GDP Change'] = np.log(df_gdp['GDPC1']).diff() * 100
df_gdp.index = pd.to_datetime(df_gdp.index)
# Convert to quarterly period and then get the quarter-end timestamp
df_gdp.index = df_gdp.index.to_period('Q').to_timestamp('Q')
df_gdp.dropna(subset=['Quarterly GDP Change'], inplace=True)
df_gdp

# Ensure all indices are in quarterly frequency and aligned
df_cpi_quarterly.index = df_cpi_quarterly.index.to_period('Q').to_timestamp('Q')
df_gdp.index = df_gdp.index.to_period('Q').to_timestamp('Q')
df_interest_quarterly.index = df_interest_quarterly.index.to_period('Q').to_timestamp('Q')
df_unemp_quarterly.index = df_unemp_quarterly.index.to_period('Q').to_timestamp('Q')

# Merge all DataFrames into a single DataFrame
df_merged = pd.concat([
    df_cpi_quarterly['Annual Inflation Rate'],
    df_gdp['Quarterly GDP Change'],
    df_interest_quarterly['Interest Rate Quarterly'],
    df_unemp_quarterly['Unemployment Rate Quarterly']
], axis=1, join='inner')

# Rename columns for clarity
df_merged.columns = ['Annual Inflation Rate', 'Quarterly GDP Change', 'Interest Rate Quarterly', 'Unemployment Rate Quarterly']

# Drop rows with missing values
df_merged.dropna(inplace=True)

# Display the merged DataFrame
print("Merged DataFrame:")
print(df_merged.head())

# Optionally, plot the merged data to visualize
df_merged.plot(subplots=True, figsize=(12, 10), title="Merged Data Visualization")
plt.tight_layout()
plt.show()
```

```python
def calculate_AR_model(df_inflation, lags):

    # Extract the 'Annual Inflation Rate' column
    inflation_rates = df_inflation['Annual Inflation Rate']

    # Fit the AR model
    model = AutoReg(inflation_rates, lags=lags)
    model_fit = model.fit()

    # Print the model summary
    display(model_fit.summary())

    # Get the AR coefficients (excluding the intercept term)
    ar_coef = model_fit.params[1:]

    # Print AR coefficients
    print("\nAR Coefficients:")
    display(ar_coef)

    # Get the residuals
    residuals = model_fit.resid

    # Plot the ACF of residuals
    plt.figure(figsize=(12, 6))
    plot_acf(residuals, lags=12, alpha=0.05)
    plt.title("Autocorrelation Function (ACF) of Residuals")
    plt.show()

    # Perform the Ljung-Box test on residuals
    lb_test = acorr_ljungbox(residuals, lags=[10], return_df=True)
    print("Ljung-Box Test Results for Residuals:")
    display(lb_test)

    # Return results
    return model_fit, ar_coef, lb_test, residuals
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.ardl import ARDL
from sklearn.metrics import mean_squared_error

# Prepare your data as before
test_data1 = df_inflation.loc[df_inflation.index >= '2023-06-01']
train_data1 = df_inflation.loc[df_inflation.index < '2023-06-01']
test_data = test_data1['Annual Inflation Rate']
train_data = train_data1['Annual Inflation Rate']
exog = df_merged[['Quarterly GDP Change', 'Interest Rate Quarterly', 'Unemployment Rate Quarterly']]
train_exog = exog.loc[exog.index < '2023-06-01']
test_exog = exog.loc[exog.index >= '2023-06-01']

# Fit the ARDL model
model_ardl = ARDL(train_data, lags=5, exog=train_exog, order=1)
result = model_ardl.fit()

# Extract coefficients
coefficients = result.params

# Manually Predicting using the coefficients
predictions = []

# Initial prediction using the last 'lag' points from the training data
# Get the last 'lags' number of data points (5 lags here)
train_lags = train_data[-5:].values
train_exog_lags = train_exog[-5:].values

# Number of exogenous variables
num_exog_vars = train_exog.shape[1]

# Loop through the test data and predict one by one
for i in range(len(test_data)):
    # Concatenate lagged values of train_data (5 lags) and exog data (5 lags for each exogenous variable)
    lags = np.concatenate([train_lags[::-1], train_exog_lags[::-1].flatten()])
    
    # Ensure lags length matches the number of coefficients
    if len(lags) != len(coefficients[1:]):
        raise ValueError(f"Mismatch in dimensions: lags length = {len(lags)}, coefficients length = {len(coefficients[1:])}")
    
    # Calculate prediction using the formula: constant + (coefficients * lags)
    y_hat = coefficients[0] + np.dot(coefficients[1:], lags)  # Intercept + lag coefficients

    # Store prediction
    predictions.append(y_hat)

    # Update train_lags and train_exog_lags with the current actual test value
    if i < len(test_data) - 1:  # Only update if not last iteration
        train_lags = np.roll(train_lags, 1)  # Shift train_lags to the right
        train_lags[0] = test_data.iloc[i]  # Replace the first element with the new prediction
        
        train_exog_lags = np.roll(train_exog_lags, 1, axis=0)  # Shift exogenous variables to the right
        train_exog_lags[0] = test_exog.iloc[i].values  # Replace with new exog data from test set

# Convert predictions into a pandas Series for easy comparison
pred_ar6_manual = pd.Series(predictions, index=test_data.index)

# Plot forecast vs actual
plt.figure(figsize=(12, 6))
plt.plot(test_data, label="Actual Inflation")
plt.plot(pred_ar6_manual, label="Manual ARDL Forecast", linestyle='--')
plt.axvline(x=test_data.index[0], color='gray', linestyle=':')
plt.title("Manual ARDL Forecast vs Actual Inflation")
plt.xlabel("Date")
plt.ylabel("Inflation")
plt.legend()
plt.tight_layout()
plt.show()

# Compute RMSE to evaluate the forecast
rmse = np.sqrt(mean_squared_error(test_data, pred_ar6_manual))
print(f"RMSE: {rmse}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error

def ar_forecast(p):
    """
    Function to fit an AR model with the specified number of lags (p), make manual predictions for the test data,
    and plot the forecast vs actual values. The train and test data are automatically split based on a predefined date.

    Args:
    - p (int): The number of lags to use for the AR model.
    
    Returns:
    - pred_ar (pd.Series): Forecasted values for the test data.
    - coef (np.ndarray): Coefficients of the AR model.
    - rmse (float): Root Mean Squared Error of the forecast.
    """
    # Assuming df_inflation is your DataFrame containing the inflation data
    test_data1 = df_inflation.loc[df_inflation.index >= '2008-06-01']
    train_data1 = df_inflation.loc[df_inflation.index < '2008-06-01']
    
    test_data = test_data1['Annual Inflation Rate']
    train_data = train_data1['Annual Inflation Rate']
    
    # 3. Fit AR(p) on train_data only
    model_ar = AutoReg(train_data, lags=p).fit()
    coef = model_ar.params  # Intercept + coefficients

    # 4. Combine full data for manual forecasting
    combined_data = pd.concat([train_data, test_data]).copy()

    # 5. Manual prediction loop using actual values
    pred_ar = []

    for i in range(len(train_data), len(combined_data)):
        lags_values = combined_data.iloc[i-p:i].values[::-1]  # lagged actual values, reversed
        y_hat = coef[0] + np.dot(coef[1:], lags_values)  # Intercept + lag coefficients
        pred_ar.append(y_hat)

    # 6. Store as a Series
    pred_ar = pd.Series(pred_ar, index=test_data.index)

    # 7. Plot forecast vs actual
    plt.figure(figsize=(12, 6))
    plt.plot(test_data, label="Actual Inflation")
    plt.plot(pred_ar, label=f"AR({p}) Forecast", linestyle='--')
    plt.axvline(x=test_data.index[0], color='gray', linestyle=':')
    plt.title(f"Manual AR({p}) Forecast vs Actual Inflation")
    plt.xlabel("Date")
    plt.ylabel("Inflation")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # 8. Calculate RMSE
    rmse = np.sqrt(mean_squared_error(test_data, pred_ar))
    print(f"RMSE for AR({p}): {rmse}")
    
    return pred_ar, coef, rmse

# Example usage:
# Call the function with a specific number of lags (e.g., 1)
# pred_ar, coef, rmse = ar_forecast(p=1)
```
